# Filtro de Imágenes - Tarea 3 - Axel Chaves Reyes, Pablo Hidalgo Navarro

```python
import cv2
import matplotlib.pyplot as plt


def aplicarFiltros(pImagen):

    # Cargar la imagen y convertirla a grises
    imagen = cv2.imread(pImagen, cv2.IMREAD_GRAYSCALE)

    # Aplicar el filtro de desenfoque gaussiano con kernel de 15x15
    imagen_desenfocada = cv2.GaussianBlur(imagen, (15, 15), 0)

    # Aplicar el filtro de Sobel con kernel 7x7
    sobel_x = cv2.Sobel(imagen, cv2.CV_64F, 1, 0, ksize=25)
    sobel_y = cv2.Sobel(imagen, cv2.CV_64F, 0, 1, ksize=25)
    sobel = cv2.magnitude(sobel_x, sobel_y)

    # Aplicar el filtro de Canny con rango 50-200
    canny = cv2.Canny(imagen, 50, 150)

    # Aplicar el filtro bilateral
    bilateral = cv2.bilateralFilter(imagen, 50, 500, 500)

    # Mostrar las imágenes
    plt.figure(figsize=(20, 10))

    plt.subplot(1, 5, 1)
    plt.imshow(imagen, cmap='gray')
    plt.title("Imagen Original")
    plt.axis("off")

    plt.subplot(1, 5, 2)
    plt.imshow(imagen_desenfocada, cmap='gray')
    plt.title("Imagen Desenfocada")
    plt.axis("off")

    plt.subplot(1, 5, 3)
    plt.imshow(sobel, cmap='gray')
    plt.title("Filtro Sobel")
    plt.axis("off")

    plt.subplot(1, 5, 4)
    plt.imshow(canny, cmap='gray')
    plt.title("Filtro Canny")
    plt.axis("off")

    plt.subplot(1, 5, 5)
    plt.imshow(bilateral, cmap='gray')
    plt.title("Filtro Bilateral")
    plt.axis("off")

    plt.show()
```

```python
aplicarFiltros("Tarea3_IA\\Dataset\\Normal_traing_050.jpeg")
```
